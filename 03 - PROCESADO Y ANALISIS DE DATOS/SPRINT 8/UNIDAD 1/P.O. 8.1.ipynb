3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Barras 
# Top 5 ciudades 
df_top5 = df.sort_values(by='population_total', ascending=False).head(5)
# Colores para ciudades 
colores = ['mediumslateblue', 'silver', 'coral', 'olive', 'indianred']
# Figura
plt.figure(figsize = (8, 6))
# Grafico
bars = plt.bar(df_top5['city'], df_top5['population_total'], color = colores, alpha = 0.7)

for bar in bars:
    poblacion = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, poblacion, f'{poblacion:,}',
             ha = 'center', va = 'bottom', color = 'black', fontsize = 10)
    
plt.xlabel('Ciudades')
plt.ylabel('Habitantes')
plt.title('Ciudades mas pobladas de California');
```

```python
# Histogramas 
plt.figure(figsize = (10, 6))
plt.hist(df['area_total_km2'], bins = 50, color = 'steelblue', alpha = 0.7)
plt.xlabel('superficie (km2)', fontname = "Bradley Hand itc", fontsize = 18)
plt.ylabel('Numero de ciudades', fontname = "Bradley Hand itc", fontsize = 18)
plt.legend(['Hist Area']);
```

```python
fig, axs = plt.subplots(2, 2, figsize = (12, 10))

# Barra de las ciudades más pobladas
df_top5 = df.sort_values(by = 'population_total', ascending = False).head(5)
colores = ['mediumslateblue', 'silver', 'coral', 'olive', 'indianred']
bars = axs[0, 0].bar(df_top5['city'], df_top5['population_total'], color=colores, alpha = 0.7)
axs[0, 0].set_xlabel('Ciudades')
axs[0, 0].set_ylabel('Habitantes')
axs[0, 0].set_title('Ciudades más pobladas de California')

for bar in bars:
    poblacion = bar.get_height()
    axs[0, 0].text(bar.get_x() + bar.get_width() / 2, poblacion, f'{poblacion:,}',
                   ha='center', va='bottom', color='black', fontsize=10)

# Histograma de áreas totales
axs[0, 1].hist(df['area_total_km2'], bins=50, color='steelblue', alpha = 0.7)
axs[0, 1].set_xlabel('Superficie (km2)', fontname="Bradley Hand itc", fontsize = 12)
axs[0, 1].set_ylabel('Numero de ciudades', fontname="Bradley Hand itc", fontsize = 12)
axs[0, 1].legend(['Hist Area'])

# Dispersión de población por área de agua
axs[1, 0].scatter(df["area_water_km2"], df["population_total"], alpha = 0.3, color = 'g')
axs[1, 0].set_xlabel("Area de agua en la ciudad (km2)")
axs[1, 0].set_ylabel("Poblacion (miles habitantes)")
axs[1, 0].set_title('Poblacion por kilometro cuadrado de agua')

# Diagrama de caja de la altitud
axs[1, 1].boxplot(df['elevation_m'], vert = False, boxprops = dict(facecolor = 'r', color = 'k'), patch_artist = True)
axs[1, 1].set_title('Diagrama de caja de la Altitud de las ciudades de California')
axs[1, 1].set_xlabel('Altitud (m)');
```
