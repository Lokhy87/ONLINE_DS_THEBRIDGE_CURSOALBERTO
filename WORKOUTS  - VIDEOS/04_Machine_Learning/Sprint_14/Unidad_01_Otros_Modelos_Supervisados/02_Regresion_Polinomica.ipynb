# Otros Modelos: Polynomial regression/Regresin Polinmica



### Concepto y modelado


En muchas ocasiones la relacin entre los datos no va a ser lineal, por lo que una simple regresin lineal no se va a ajustar bien a nuestra problemtica.

Por ejemplo, en el siguiente ejemplo encajaria muy bien una regresin lineal.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline 

df = pd.read_csv('data/salary_data.csv')
df.head()
```

```python
sns.regplot(data=df,  # regplot una nueva herramienta visual para tu caja de herramientas
            x="YearsExperience",
            y="Salary");
```

En la vida real los problemas son ms complejos, ya que los salarios no llevan una subida lineal vs aos de experiencia, depender tambin de la carrera profesional: profesor, ingeniero, doctor...

Otra posible representacin del salario podra ser en funcin del puesto de trabajo.

```python
df = pd.read_csv('data/position_salaries.csv')
df
```

Vamos a construir un modelo que tenga en cuenta el "nivel" profesional

```python
X = df.iloc[:, 1:2].values
y = df.iloc[:, 2].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Visualizing the Linear Regression results
def viz_linear():
    plt.scatter(X, y, color='red')
    plt.plot(X, lin_reg.predict(X), color='blue')
    plt.title('Verdad o Fracaso (Linear Regression)')
    plt.xlabel('Nivel Profesional')
    plt.ylabel('Salario')
    plt.show()
    return
viz_linear()
```

Vemos que en este caso, la regresin lineal no se ajusta nada bien a los datos, por lo que habr que probar otras alternativas. Observa los errores:

```python
plt.scatter(X, y - lin_reg.predict(X));
```

Y aqu es donde entra la **regresin polinmica** que consiste en aplicar una transformacin a nuestra $y = a_0 + a_1*x_1 + a_2*x_2 + a_3*x_3 + ...$ (donde $x_i$ son las features y $a_i$ los coeficientes o pesos o parmetros de una regresin lineal) sencilla pero potente. En vez de esa relacin vamos a utilizar tambin los cuadrados, cubos, etc de las mismas features y de los productos de las mismas:

$$y = a_0 + a_1*x_1 + a_2*x_1^2 + a_3*(x_1*x_2) + a_4*x_2^2 + a5*x_3^2 + a_6*(x_1*x_3) + a_7*(x_2*x_3) + a_8*x_1^3...$$

El grado de la regresin polinmica es el mximo exponente al que elevaremos las features (en el caso anterior sera 3). Por otro lado, fjate que es como crear una regresin lineal incluyendo las features "creadas" o "sintticas" ($x_1^2, x_2^2, x_3^2, x_1^3, x_2^3, x_3^3, x_1*x_2,x_1*x_3,x_2*x_3,x_1*x_2^2,...$, etc)

**

Por supuesto, `sklearn` tiene su clase que se encarga de obtener esas features "sintticas" y luego nosotros haremos la regresin lineal con todo el nuevo conjunto de datos. Eso s, tendrs que indicarle el grado :-) a travs del argumento `degree`. La clase es `PolynomialFeatures`. Veamos cmo, pero antes observa que tambin crea los productos combinados de features:

```python
X
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=3)
poly_reg.fit(X)
X_poly = poly_reg.transform(X)
X_poly
```

Como solo teniamos una feature ($x_1$) crear un nuevo dataset con:

$$(x_1^0, x_1^1, x_1^2, x_1^3)$$

Ahora ejecutamos el modelado y veamos la comparativa:

```python
# Linear regression
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)

# Visualizing the Polymonial Regression results
def viz_polymonial():
    plt.scatter(X, y, color='red')
    plt.plot(X, pol_reg.predict(poly_reg.transform(X)), color='blue')
    plt.title('Truth or Bluff (Linear Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
    return
viz_polymonial()
```

### Sobreajuste

El problema es que si aadimos mucha complejidad (empleamos grados muy altos) podemos sobreajustar el modelo fcilmente. Por ejemplo ejecutemos la siguiente celda (en la que se crean unos datos sintticos y luego se ajusta un modelo de regresin polinmica forzado para el sobreajuste)

```python
# Paso 1: Generar datos sintticos
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # 100 puntos aleatorios entre 0 y 10
y = np.sin(X).ravel() + np.random.normal(0, 0.2, X.shape[0])  # y = sin(x) con ruido
plt.scatter(X,y);
```

```python
# Paso 2: Ajustar un modelo de regresin polinmica de alto grado
grado = 15  # Grado del polinomio
poly_overfitting = PolynomialFeatures(degree = grado)
X_poly = poly_overfitting.fit_transform(X)
modelo = LinearRegression()
modelo.fit(X_poly, y)
```

```python
# Paso 3: Visualizar los resultados
X_test = np.linspace(0, 10, 100)[:, np.newaxis]
X_test_poly = poly_overfitting.transform(np.linspace(0, 10, 100)[:, np.newaxis])
y_pred = modelo.predict(X_test_poly)

plt.scatter(X, y, label="Datos originales")
plt.plot(X_test, y_pred, color='red', label=f"Regresin polinmica de grado {grado}")
plt.legend()
plt.title("Sobreajuste con Regresin Polinmica")
plt.show()
```

El ajuste es casi perfecto, lo cual quiere decir que en la realidad con datos nuevos que cambiarn algo el patrn el modelo dejar de generalizar perfectamente (olvida por un momento que hemos hecho los datos sintticos)

### Predicciones

Y para terminar, para obtener las predicciones de un nuevo valor (o de un conjunto de test), observa que en realidad hemos aplicado una transformacin al train (como hacemos cuando aplicamos logaritmos para normalizar algunas distribuciones) y por tanto lo que hay que hacer es aplicar dicha tranformacin a los datos de entrada.

```python
# Predicting a new result with Linear Regression
print(lin_reg.predict([[5.5]]))

# Predicting a new result with Polymonial Regression
print(pol_reg.predict(poly_reg.transform([[5.5]])))
```
