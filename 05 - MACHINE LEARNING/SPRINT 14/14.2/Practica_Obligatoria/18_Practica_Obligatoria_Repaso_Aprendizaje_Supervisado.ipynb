### #1 Explicacin del dataset y carga de datos

Vamos a trabajar con el dataset "Wine Quality", que es un dataset en el que se recogen diversas caractersticas fisicoqumicas de las variades tinta y blanca del "Vinho verde" portugus. Las variables del dataset se describen a continuacin


1. **Acidez fija**: La mayora de los cidos involucrados con el vino son fijos o no voltiles (no se evaporan fcilmente).
2. **Acidez voltil**: La cantidad de cido actico en el vino, que en niveles demasiado altos puede llevar a un sabor desagradable a vinagre.
3. **cido ctrico**: Encontrado en pequeas cantidades, el cido ctrico puede aadir 'frescura' y sabor a los vinos.
4. **Azcar residual**: Es la cantidad de azcar restante despus de que se detiene la fermentacin, es raro encontrar vinos con menos de 1 gramo/litro y vinos con ms de 45 gramos/litro se consideran dulces.
5. **Cloruros**: La cantidad de sal en el vino.
6. **Dixido de azufre libre**: La porcin libre de SO2 existe en equilibrio entre el SO2 molecular (como un gas disuelto) y el ion bisulfito; previene el crecimiento microbiano y la oxidacin del vino.
7. **Dixido de azufre total**: Es la cantidad de formas libres y ligadas de S02; en bajas concentraciones, el SO2 es en su mayora indetectable en el vino, pero en concentraciones libres de ms de 50 ppm, el SO2 se hace evidente en la nariz y el sabor del vino.
8. **Densidad**: La densidad del vino es cercana a la del agua dependiendo del porcentaje de alcohol y contenido de azcar.
9. **pH**: Describe qu tan cido o bsico es un vino en una escala de 0 (muy cido) a 14 (muy bsico); la mayora de los vinos estn entre 3-4 en la escala de pH.
10. **Sulfatos**: Un aditivo del vino que puede contribuir a los niveles de dixido de azufre (SO2), que acta como un antimicrobiano y antioxidante.
11. **Alcohol**: El porcentaje de contenido de alcohol del vino.
12. **Calidad**: Una variable de salida (basada en datos sensoriales, puntuacin entre 0 y 10).
13. **Class**: Si es blanco o tinto.

```python
features_num_clf = ['chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'alcohol']
```

```python
# Variable categorica en binaria

train_set['class'] = (train_set['class'] == 'white').astype(int)
test_set['class'] = (test_set['class'] == 'white').astype(int)
```

```python
features_num_clf
```

```python
train_set[features_num_clf].hist(figsize = (15, 10));
```

```python
train_set_scaled = train_set.copy()
test_set_scaled = test_set.copy()
```

```python
features_to_transform = ['chlorides', 'free sulfur dioxide', 'total sulfur dioxide']
```

```python
# Transformacion logaritmica

epsilon = 1e-9

for feature in features_to_transform:
    train_set_scaled[feature] = np.log(train_set_scaled[feature] + epsilon)
    test_set_scaled[feature] = np.log(test_set_scaled[feature] + epsilon)
```

```python
# Escalado

scaler = StandardScaler()
features_num_reg = features_num_clf.copy()
train_set_scaled[features_num_reg] = scaler.fit_transform(train_set_scaled[features_num_reg])
test_set_scaled[features_num_reg] = scaler.fit_transform(test_set_scaled[features_num_reg])
```

```python
train_set_scaled_clf = train_set_scaled.copy()
test_set_scaled_clf = test_set_scaled.copy()

train_set_scaled_clf['alcohol'] = scaler.fit_transform(train_set_scaled_clf[['alcohol']])
test_set_scaled_clf['alcohol'] = scaler.transform(test_set_scaled_clf[['alcohol']])
```

```python
# X - Y

X_train = train_set_scaled_clf[features_num_clf]
y_train = train_set_scaled_clf[target_clf]

X_test = test_set_scaled_clf[features_num_clf]
y_test = test_set_scaled_clf[target_clf]
```

KNN

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
```

```python
k_values = [3, 5]

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_imputed, y_train)
    
    y_pred = knn.predict(X_test_imputed)
    
    print(f"Classification Report for KNN (k={k}):")
    print(classification_report(y_test, y_pred))
```

RandomForest

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest:")
print(classification_report(y_test, y_pred_rf))
```

```python
param_grid_rf = {
    'n_estimators': [50, 100, 200],  # Nmero de rboles en el bosque
    'max_depth': [None, 10, 20],       # Profundidad mxima de cada rbol
    'min_samples_split': [2, 5, 10],   # Nmero mnimo de muestras requeridas para dividir un nodo interno
    'min_samples_leaf': [1, 2, 4]      # Nmero mnimo de muestras requeridas en un nodo hoja
}

rf_grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring='recall_macro')
rf_grid_search.fit(X_train_imputed, y_train)
```

```python
best_params_rf = rf_grid_search.best_params_
best_rf_model = rf_grid_search.best_estimator_

y_pred_best_rf = best_rf_model.predict(X_test_imputed)
```

```python
print(f"Best Parameters: {best_params_rf}")
print(classification_report(y_test, y_pred_best_rf))
```
