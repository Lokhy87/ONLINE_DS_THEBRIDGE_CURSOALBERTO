Modelo

**DBSCAN**

```python
X = df_energ[selected_features]
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps = 0.5,
                min_samples = 5)
dbscan.fit(X)
```

```python
# Datos etiquetados 
print(len(dbscan.labels_))

# 10 primeros labels 
print(dbscan.labels_[:10])

# Clusters montados
print(np.unique(dbscan.labels_))
```

```python
# Cores_sample_indices
print(len(dbscan.core_sample_indices_))

# Indices de 10 primeros core_sample
print(dbscan.core_sample_indices_[:10])

# Posicion 5 primeros core_sample
print(dbscan.components_[:5])
```

```python
labels = dbscan.labels_

# Crear una mscara para los core samples
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

# Asignar diferentes colores a los clusters y al ruido
unique_labels = np.unique(labels)
colors = plt.cm.get_cmap('plasma', len(unique_labels))

# Plotear los clusters
plt.figure(figsize=(12, 8))
for k in unique_labels:
    if k == -1:
        # El color negro se usa para el ruido
        col = 'k'
        label_name = 'Ruido'
    else:
        col = colors(k)
        label_name = f'Cluster {k}'

    # Crear una mscara para cada cluster
    class_member_mask = (labels == k)

    # Plotear los core samples
    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy.iloc[:, 0], xy.iloc[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14, label=label_name)

    # Plotear los puntos de borde
    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy.iloc[:, 0], xy.iloc[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=6)

plt.title('Clusters obtenidos por DBSCAN')
plt.xlabel('Cafena')
plt.ylabel('Taurina')
plt.legend(loc='best')
plt.show()

```
